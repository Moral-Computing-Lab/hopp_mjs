# Boostrapping the MJS

In this notebook we bootstrap the MJS to obtain a map of the voxels that most reliably predict moral wrongness ratings.

```python
import glob
import pandas as pd
from nltools.data import Brain_Data, Design_Matrix

from nltools.stats import (zscore, 
                           regress, 
                           find_spikes, 
                           one_sample_permutation, 
                           _calc_pvalue, 
                           downsample, 
                           correlation_permutation, 
                           summarize_bootstrap, 
                           threshold, 
                           fdr
                          )
```

```python
mjs_boot = Brain_Data('weightmaps/mjs_bootstrap_n10000.hdf5')
```

```python
mjs_boot_summary = summarize_bootstrap(mjs_boot)
```

```python
mjs_boot_summary['Z'].write('weightmaps/mjs_bootstrap_Z_n10000.hdf5')
mjs_boot_summary['p'].write('weightmaps/mjs_bootstrap_p_n10000.hdf5')
```

```python
bootstrap_z = mjs_boot_summary['Z']
bootstrap_p = mjs_boot_summary['p']

fdr_thresh = fdr(bootstrap_p.data, q=0.05)
mjs_thresholded = threshold(bootstrap_z, bootstrap_p, thr=fdr_thresh)
mjs_thresholded.write('weightmaps/mjs_bootstrap_thresh_n10000.hdf5')
```
